```python
# Pulling data from the BOC to build macroeconomic charts
# Resource: https://www.askpython.com/python/examples/pull-data-from-an-api

# BOC API pulls begin with https://www.bankofcanada.ca/valet/
# INDEX OF API CODES
# Total CPI = v41690973 
# Total CPI (seasonally adjusted) = v41690914 
# Total CPI change over last 12 months = v41690973


import requests

response_api = requests.get('https://www.bankofcanada.ca/valet/observations/v41690973/csv')
data = response_api.text
print(response_api)
print(data)
```
